```python
import os
import pickle

directory = '/home/thattai/kritikak/Param_Brahma_IISERP/Mk_array_job/4c13e/chunks/'
output_directory = '/home/thattai/kritikak/Param_Brahma_IISERP/Mk_array_job/4c13e/chunks_real/'

if not os.path.exists(output_directory):
    os.makedirs(output_directory)

def split_list_into_chunks(lst, chunk_size):
    for i in range(0, len(lst), chunk_size):
        yield lst[i:i+chunk_size]

# Iterate over files in the directory
for filename in os.listdir(directory):
    if filename.startswith('Res_'):
        input_path = os.path.join(directory, filename)
        output_filename = os.path.splitext(filename)[0]
        # output_dir = os.path.join(output_directory, os.path.splitext(filename)[0])

        # Create a new directory for each file
        # if not os.path.exists(output_dir):
        #     os.makedirs(output_dir)

        # output_path = os.path.join(output_directory, output_filename)

        with open(input_path, 'rb') as f:
            original_data = pickle.load(f)

        chunk_size = 5
        data_chunks = list(split_list_into_chunks(original_data, chunk_size))

        for i, chunk in enumerate(data_chunks):
            with open(f'{output_directory}{i+1}_{output_filename}.pkl', 'wb') as f:
                pickle.dump(chunk, f)
```
